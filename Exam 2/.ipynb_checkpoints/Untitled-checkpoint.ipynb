```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x_values = [10, 20, 30]
y_values = [20, 40, 10]
ax.plot(x_values, y_values,':b', label='line1-dotted')

x2_values = [10, 20, 30]
y2_values = [40,10,30]
ax.plot(x2_values, y2_values, '--r', label='line2-dashed')

leg = ax.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=50)
y = np.random.randint(100, size=50)

plt.scatter(x,y, color='g')

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ax = plt.axes()

x = np.arange(6)
y = 5 * x

plt.title("y = 5x line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
ax.plot(x, y, c="g")
plt.show()
```
